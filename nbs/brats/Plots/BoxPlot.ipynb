```python
def create_new_df(df): 
    new = df[['dice']].copy()
    return new
```

```python
df_0 = create_new_df(df_0)['dice']
df_1 = create_new_df(df_1)['dice']
df_2 = create_new_df(df_2)['dice']
df_3 = create_new_df(df_3)['dice']
df_4 = create_new_df(df_4)['dice']
df_5 = create_new_df(df_5)['dice']
```

```python
frames = [df_0, df_1, df_2, df_3, df_4, df_5]
data = np.array(frames).T
```
